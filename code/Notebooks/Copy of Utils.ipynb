```python
df = pd.read_csv("/gdrive/Shareddrives/FYP Project/Dataset/FoodData_modified.csv")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
```

```python
from Utils import preprocess_dataset
```

```python
data = preprocess_dataset(df)
```

```python
data[3]
```

```python
X=data[0][["Calories (kcal)",'Protein (g)', 'Carbs (g)', 'Fats (g)']].values
```

```python
X
```

```python
Y=data[2]
```

```python
x_train, x_test, y_train, y_test = train_test_split( X, Y, test_size=0.33, random_state=42)
```

```python
clf =  RandomForestClassifier(n_estimators = 100, max_depth = 29, max_features = 4, max_samples = 1.0, random_state=0)
clf.fit(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

```python
clf.predict(np.array([400, 0, 99.6, 0]).reshape(1,-1))
```

```python
history = {"clf": clf.score(x_test, y_test)}
```

```python
pipeline = Pipeline(steps=[("imputer", SimpleImputer(strategy="constant")),("model", RandomForestClassifier(n_estimators = 100, max_depth = 29, max_features = 4, max_samples = 1.0, random_state=0)) ])
```

```python
pipeline.fit(x_train, y_train)
```

```python
pipeline.score(x_test, y_test)
```

```python
pipeline.predict(np.array([77, 2, 17, 0.1]).reshape(1,-1))
```
