```python
# Category 0
df0 =pd.read_csv(r'/gdrive/Shareddrives/FYP Project/Dataset_October25_Clustering/FoodData_group0.csv')
# Category 1
df1 =pd.read_csv(r'/gdrive/Shareddrives/FYP Project/Dataset_October25_Clustering/FoodData_group1.csv')
# Category 2
df2 =pd.read_csv(r'/gdrive/Shareddrives/FYP Project/Dataset_October25_Clustering/FoodData_group2.csv')
# Category 3
df3 =pd.read_csv(r'/gdrive/Shareddrives/FYP Project/Dataset_October25_Clustering/FoodData_group3.csv')
```

```python
# Combine all categories
df_all = pd.concat([df0, df1, df2, df3])
```

```python
# Create a copy of data set
df_std = df_all.copy()
```

```python
df_std = df_std.drop(["Unnamed: 0"], axis=1)
```

```python
def plot_graph(category):
  if type(category) is str:
    subcategory = df_std[df_std["Group"]==category]
    title="%s Nutrients average vs categories"%(category.capitalize())
  else:
    subcategory = df_std[df_std["Group"].isin(category)]
    title = f"{' & '.join(category).capitalize()} Nutrients average vs categories"
  source = subcategory.groupby("Category").mean()
  plot_bar = source.plot_bokeh.bar(ylabel="Nutrients range", title=title, legend = "top_right", figsize=(500, 200), sizing_mode="scale_width")
```
